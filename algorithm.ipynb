```python
import math
from numpy.linalg import inv
import numpy as np
import matplotlib.pyplot as plt


# Find the square of distance 
def distSq(a,b):
    return ((a[0]-b[0])**2 + (a[1]-b[1])**2)

def point(prev,pres,latlong):
    
#     inv |p1.x − p0.x, p1.y − p0.y|   |q.x(p1.x − p0.x) + q.y(p1.y − p0.y) |
#         |p0.y − p1.y, p1.x − p0.x| . |p0.y(p1.x − p0.x)- p0.x(p1.y − p0.y)|
    try:
        ainv = inv(np.array([[pres[0]-prev[0],pres[1]-prev[1] ], [prev[1]-pres[1], pres[0]-prev[0]]]))
        b = np.array([[latlong[0]*(pres[0]-prev[0])+latlong[1]*(pres[1]-prev[1])],[prev[1]*(pres[0]-prev[0])-prev[0]*(pres[1]-prev[1])]])
        point = [ainv[0][0]*b[0] + ainv[0][1]*b[1] , ainv[1][0]*b[0] + ainv[1][1]*b[1] ] 
        m1 = (pres[1]-prev[1])/(pres[0]-prev[0])
        m2 = (point[1]-latlong[1])/(point[0]-latlong[0])
#         print "Slope",m1*m2
        if(point != latlong):
#             print "wrong"
            print str(point[0]) + ' ---- ' + str(latlong[0])
#         plt.plot([prev[0],pres[0]],[prev[1],pres[1]],linestyle='-', marker='o')
#         plt.plot(latlong[0],latlong[1],marker='o')
#         plt.plot(point[0],point[1],marker='o')
#         plt.show()
        return [point[0],point[1]]
    except:
        pass
    return [-1,-1] 
    
```

```python
def isAcute(prev,pres,latlong):
    a = distSq(prev,pres)
    b = distSq(latlong,pres)
    c = distSq(prev,latlong)
    if(a>b):
        if(b>c):
            #a is the largest
            if(a<=b+c):
                #accute 
                return True
            else:
                return False
        else: #(c>b)
            if(c>a):
                #c is the largest
                if(c<=a+c):
                    #accute 
                    return True
                else:
                    return False
            else:
                #b<c<a
                #a is the largest
                if(a<=b+c):
                    #accute 
                    return True
                else:
                    return False
    else: #b>a
        if(a>c):
            #b is the largest
            if(c<=a+c):
                #accute 
                return True
            else:
                return False
        else:
            #b>a<c
            if(b>c):
                #b is the largest
                if(c<=a+c):
                    #accute 
                    return True
                else:
                    return False                
            else:
                #c is the largest
                if(c<=a+c):
                    #accute 
                    return True
                else:
                    return False


def findNearestPoint(latlong):
#     latlong = [19.127300000,72.916220000]
    pres = []
    prev = []
    final= []
    flag = 0
    min1 =1000
#     print latlong
    f = open('latlong.txt','r')
    for i in f:
        j= i.split()
        if (i=='\n'): #new route 
            flag =2
        else : #same route => update
            if(flag==2):
                prev = [float(j[1]),float(j[2])]
                flag = 1
#                 print "prev"
            if(flag==1):
                pres = [float(j[1]),float(j[2])]
#                 print "pres"
                flag = 0
                if(isAcute(prev,pres,latlong)):
                    p = point(prev,pres,latlong)
                    d = distSq(p,latlong)
                    if(d<min1):
                        min1 = d
                        final = p
                
            if(flag==0):
                prev = pres
#                 print "c"
                pres= float(j[1]),float(j[2])
                if(isAcute(prev,pres,latlong)==True):
                    p = point(prev,pres,latlong)
                    if(p[0]!=-1):
                        d = distSq(p,latlong)
                        if(d<min1):
                            min1 = d
                            final = p
                        
    return final
```

```python
import csv
f = open("sample.txt")
fo = open("outlatlong.csv",'w')
w = csv.writer(fo)
# w.writerow(['lat','long','TimeStamp'])
for i in f:
    l = i.split()
    time = l[1]+' '+l[2]
    p = [float(l[3]),float(l[4])]
    final = findNearestPoint(p)
    if(final!=[]):
        final = [final[0][0],final[1][0]]
        final.append(time)
    else:
        final = p
        final.append(time)
#     if (final == p):
#         final.append('old')
    w.writerow(final)
#     fo.writeline(str())

    
```
