```python
import math
from numpy.linalg import inv
import numpy as np
import matplotlib.pyplot as plt


# Find the square of distance 
def distSq(a,b):
    return ((a[0]-b[0])**2 + (a[1]-b[1])**2)

def point(prev,pres,latlong):
#     inv |p1.x − p0.x, p1.y − p0.y|   |q.x(p1.x − p0.x) + q.y(p1.y − p0.y) |
#         |p0.y − p1.y, p1.x − p0.x| . |p0.y(p1.x − p0.x)- p0.x(p1.y − p0.y)|
    ainv = inv(np.array([[pres[0]-prev[0],pres[1]-prev[1] ], [prev[1]-pres[1], pres[0]-prev[0]]]))
    b = np.array([[latlong[0]*(pres[0]-prev[0])+latlong[1]*(pres[1]-prev[1])],[prev[1]*(pres[0]-prev[0])-prev[0]*(pres[1]-prev[1])]])
    point = [ainv[0][0]*b[0] + ainv[0][1]*b[1] , ainv[1][0]*b[0] + ainv[1][1]*b[1] ] 
    plt.plot([prev[0],pres[0]],[prev[1],pres[1]],linestyle='-', marker='o')
    plt.plot(latlong[0],latlong[1],marker='o')
    plt.plot(point[0],point[1],marker='o')
    plt.show()
    
    return point 
    
```
